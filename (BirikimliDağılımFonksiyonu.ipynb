```python
from scipy import stats
import  matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
olasılık = stats.norm.pdf(x = 11,loc = 15 ,scale = 3)
print(olasılık)
```

f(11) in değeri 0.54 yani alan değil.
Alan hesabı içi CDF kullanırılır o değere gelene kadar tüm değerleri toplar ve alan bulunur.
F(x) = kümilatif olasılık yapısı

```python
olasılık2 = stats.norm.cdf(x = 11,loc = 15 ,scale = 3)
print(olasılık2)
```

```python
x = [0,1,2,3]
p = [1/8,3/8,3/8,1/8]
cum =[]

for i in range(0,len(p)):
    if len(cum) == 0:
        cum.append(p[0])
    else:
        cum.append(p[i] + cum[i-1]) 

print(cum)
```

```python
plt.plot(x,cum,drawstyle = "steps")
plt.show()
```

```python
x = np.random.randn(10000)
pdf = stats.norm.pdf(x)
cdf = stats.norm.cdf(x)
sns.lineplot(x = x, y = pdf)
sns.lineplot(x = x, y = cdf)
plt.show()
```
